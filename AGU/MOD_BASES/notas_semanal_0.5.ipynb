```python
import pandas as pd
import openpyxl
import numpy as np
```

```python
# carregando a base de dados
notas_semanal_5 = pd.read_excel('notas_semanal_0.5.xlsx', header=2)

# removendo os espaaços desnecessários da base de dados
notas_semanal_5.columns = notas_semanal_5.columns.str.strip()

display(notas_semanal_5.head(5))
notas_semanal_5.info()
```

```python
# excluido colunas desncessárias
notas_semanal_5.drop(columns=['NS - Tipo Título Crédito', 'NS - Valor Linha', 'NS - Inscrição 1', 'Unnamed: 13', 'Doc - Lançado por', 'Título Crédito'], inplace=True)

display(notas_semanal_5.head(5))
notas_semanal_5.info()
```

```python
# renomeando as colunas
notas_semanal_5.rename(columns={
    'NS - Título Crédito': 'notas_documento_habil',
    'Favorecido Doc.': 'notas_doc_fornecedor',
    'Unnamed: 2': 'notas_nome_fornecedor',
    # 'NS - Inscrição 1': 'notas_empenho',
    'NS - Mês Valorização': 'notas_mes_valorizacao',
    'NS - Dia Valorização': 'notas_data_valorizacao',
    'Doc - Observação': 'notas_doc_observacao',
    'Título Crédito': 'notas_empenho',
    'Doc - Lançado por': 'notas_id_setor',
    'Unnamed: 11': 'notas_nome_servidor',
    'Unnamed: 12': 'notas_cod_servidor'
    #'Unnamed: 13': 'notas_nome_servidor',
}, inplace=True)

display(notas_semanal_5.head(5))
notas_semanal_5.info()
```

```python
# removendo colunas vazias
notas_semanal_5.dropna(axis=1, how='all', inplace=True)

# removendo linhas inválidas
notas_semanal_5 = notas_semanal_5[notas_semanal_5['notas_documento_habil'].notna()]
notas_semanal_5 = notas_semanal_5[notas_semanal_5['notas_documento_habil'] != "'-8"]
#notas_semanal_5 = notas_semanal_5[notas_semanal_5['notas_empenho'].notna()]
#notas_semanal_5 = notas_semanal_5[notas_semanal_5['notas_empenho'] != "'-8"]

# excluindo os 11 primeiros caracteres da coluna 'notas_documento_habil'
# notas_semanal_5['notas_documento_habil'] = notas_semanal_5['notas_documento_habil'].str[11:]
# notas_semanal_5['notas_empenho'] = notas_semanal_5['notas_empenho'].str[11:]

# converter a coluna 'notas_data_valorizacao' para o formato datetime
notas_semanal_5['notas_data_valorizacao'] = pd.to_datetime(notas_semanal_5['notas_data_valorizacao'], errors='coerce')

display(notas_semanal_5.head(5))
notas_semanal_5.info()
```

```python
# criando chave de relacionamento
#notas_semanal_5['notas_chave_relacionamento'] = notas_semanal_5['notas_documento_habil'].astype(str) + '_' + notas_semanal_5['notas_doc_fornecedor'].astype(str)
```

```python
# removendo duplicados
notas_semanal_5.drop_duplicates(inplace=True)
#notas_semanal_5.drop_duplicates(subset=['notas_chave_relacionamento'], inplace=True)

display(notas_semanal_5.head(10))
notas_semanal_5.info()
```

```python
# salvando a base de dados tratada
notas_semanal_5.to_excel('notas_semanal_0.5_tratada.xlsx', sheet_name='Notas_semanal_5', index=False)
```
