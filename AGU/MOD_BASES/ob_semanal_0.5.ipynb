```python
import pandas as pd
import openpyxl
import numpy as np
import datetime
```

```python
# carregando a base de dados
ob_semanal_5 = pd.read_excel('ob_semanal_0.5.xlsx', header=2)

# removendo os espaaços desnecessários da base de dados
ob_semanal_5.columns = ob_semanal_5.columns.str.strip()

display(ob_semanal_5.head(2))
ob_semanal_5.info()
```

```python
# excluindo colunas desnecessárias
ob_semanal_5.drop(columns=['Unnamed: 10'], inplace=True)

display(ob_semanal_5.head(2))
ob_semanal_5.info()
```

```python
ob_semanal_5.rename(columns = {
    'Documento Hábil': 'ob_documento_habil',
    'Compr. - Dia Pagamento': 'ob_data_pagamento',
    'Emissão - Mês': 'ob_mes_emissao',
    'OB': 'ob_ordem_pagamento',
    'Favorecido Doc.': 'ob_doc_fornecedor',
    'Unnamed: 5': 'ob_nome_fornecedor',
    'OB - Processo': 'ob_processo',
    'Doc - Observação': 'ob_observacao',
    'Doc - Valor': 'ob_valor',
    'OB - Evento': 'ob_evento',
    'OB - Inscrição 1': 'ob_empenho',

}, inplace=True)

display(ob_semanal_5.head(5))
ob_semanal_5.info()
```

```python
# removendo colunas vazias
ob_semanal_5.dropna(axis=1, how='all', inplace=True)

# removendo linhas inválidas
ob_semanal_5 = ob_semanal_5[ob_semanal_5['ob_documento_habil'].notna()]
ob_semanal_5 = ob_semanal_5[ob_semanal_5['ob_documento_habil'] != "'-8"]
ob_semanal_5 = ob_semanal_5[ob_semanal_5['ob_empenho'].notna()]
ob_semanal_5 = ob_semanal_5[ob_semanal_5['ob_empenho'] != "'-8"]

# excluindo os 11 primeiros caracteres da coluna 'ob_documento_habil'
ob_semanal_5['ob_documento_habil'] = ob_semanal_5['ob_documento_habil'].str[11:]
#ob_semanal_5['ob_empenho'] = ob_semanal_5['ob_empenho'].str[11:]

# converter a coluna 'ob_data_pagamento' para o formato datetime
ob_semanal_5['ob_data_pagamento'] = pd.to_datetime(ob_semanal_5['ob_data_pagamento'], errors='coerce')

# convertendo a coluna 'ob_evento' para string
ob_semanal_5['ob_evento'] = ob_semanal_5['ob_evento'].astype(str)

# convertendo a coluna 'ob_valor' para tipo float
ob_semanal_5['ob_valor'] = pd.to_numeric(
    ob_semanal_5['ob_valor']
    .astype(str)
    .str.replace('R$', '', regex=False)
    .str.replace('.', '', regex=False)
    .str.replace(',', '.', regex=False)
    .str.strip(),
    errors='coerce'
)

display(ob_semanal_5.head(2))
ob_semanal_5.info()
```

```python
# removendo duplicados
ob_semanal_5.drop_duplicates(inplace=True)
#ob_semanal_5.drop_duplicates(subset=['dh_chave_relacionamento'], inplace=True)

display(ob_semanal_5.head(2))
ob_semanal_5.info()
```
