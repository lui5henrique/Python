```python
import pandas as pd
import openpyxl
import numpy as np
```

```python
# carregando a base de dados
dh_semanal_5 = pd.read_excel('dh_semanal_0.5.xlsx', header=2)

# removendo os espaaços desnecessários da base de dados
dh_semanal_5.columns = dh_semanal_5.columns.str.strip()

display(dh_semanal_5.head(2))
dh_semanal_5.info()
```

```python
# excluido colunas desncessárias
dh_semanal_5.drop(columns=['DH - Doc. Origem', 'DetaCusto DH - R$'], inplace=True)

display(dh_semanal_5.head(2))
dh_semanal_5.info()
```

```python
# renomeando as colunas
dh_semanal_5.rename(columns={
    'Documento Hábil': 'dh_documento_habil',
    'DH - Credor': 'dh_doc_fornecedor',
    'Unnamed: 3': 'dh_nome_fornecedor',
    'DH - Mês Emissão': 'dh_mes_emissao',
    'DH - Valor': 'dh_valor',
    'DH - Observação': 'dh_observacao',
    'DH Item - Nota Empenho': 'dh_empenho'
}, inplace=True)

display(dh_semanal_5.head(2))
dh_semanal_5.info()
```

```python
# removendo colunas vazias
dh_semanal_5.dropna(axis=1, how='all', inplace=True)

# removendo linhas inválidas
dh_semanal_5 = dh_semanal_5[dh_semanal_5['dh_documento_habil'].notna()]
dh_semanal_5 = dh_semanal_5[dh_semanal_5['dh_empenho'].notna()]
dh_semanal_5 = dh_semanal_5[dh_semanal_5['dh_empenho'] != "'-8"]

# excluindo os 11 primeiros caracteres da coluna 'dh_documento_habil'
dh_semanal_5['dh_documento_habil'] = dh_semanal_5['dh_documento_habil'].str[11:]
dh_semanal_5['dh_empenho'] = dh_semanal_5['dh_empenho'].str[11:]


display(dh_semanal_5.head(2))
dh_semanal_5.info()
```

```python
# removendo duplicados
dh_semanal_5.drop_duplicates(inplace=True)
#dh_semanal_5.drop_duplicates(subset=['dh_chave_relacionamento'], inplace=True)

display(dh_semanal_5.head(2))
dh_semanal_5.info()
```

```python
# salvando a base de dados tratada
dh_semanal_5.to_excel('dh_semanal_0.5_tratada.xlsx', sheet_name='DH_semanal_5', index=False)
```
