```python
# Importação das Bibliotecas
import pandas as pd # para manipulação de dados
import os # para manipulação de arquivos e diretórios
import shutil # para mover arquivos entre diretórios
import datetime # para manipulação de datas
import calendar # para manipulação de calendários
import numpy as np # para manipulação de arrays e operações matemáticas
```

```python
# mover um arquivo de uma pasta para outra e renomeá-lo
pasta_origem = r'C:\Users\luis.henrique\Downloads' # caminho do diretório onde está o arquivo Excel
arquivo_excel = 'Empenhos  Contratos.gov.br.xlsx' # nome do arquivo Excel que se deseja ler
pasta_destino = r'C:\Users\luis.henrique\AGU\FileServer-SADPE - Documentos\POWER_BI\Empenhos'  # caminho completo do arquivo Excel
novo_nome_arquivo = 'base_empenhos_2026_0.1_tratada.xlsx' # novo nome do arquivo Excel
caminho_origem = os.path.join(pasta_origem, arquivo_excel) # caminho completo do arquivo Excel de origem
caminho_destino = os.path.join(pasta_destino, novo_nome_arquivo) # caminho completo do arquivo Excel de destino com o novo nome

# verificar se o arquivo Excel existe no diretório de origem
if os.path.exists(caminho_origem):
    shutil.move(caminho_origem, caminho_destino) # mover e renomear o arquivo Excel

# remover da pasta_origem o arquivo Excel
if os.path.exists(caminho_origem):
    os.remove(caminho_origem)
```

```python
# Carregamento da base de Dados - Empenhos
base_empenhos = pd.read_excel('base_empenhos_2026_0.1_tratada.xlsx', header=1)
#base_empenhos = pd.read_excel('base_empenhos_2026_0.1.xlsx', header=0)
display(base_empenhos.head(2))
base_empenhos.info()

# excluindo todos os espaços desnecessários do início e do fim dos valores das colunas
base_empenhos = base_empenhos.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
# Exclusão de colunas desnecessárias
colunas_excluir = ['Atualizado Em']
base_empenhos = base_empenhos.drop(columns=colunas_excluir)

display(base_empenhos.head(2))
base_empenhos.info()
```

```python
# mudar nome das colunas para facilitar a manipulação
base_empenhos = base_empenhos.rename(columns=
                                       {'Número do instrumento': 'empenhos_num_instrumento',
                                        'Fornecedor': 'empenhos_doc_fornecedor',
                                        'Objeto': 'empenhos_objeto',
                                        'Número Empenho': 'empenhos_num_empenho',
                                        'Plano Interno': 'empenhos_plano_interno',
                                        'Empenhado': 'empenhos_R$_empenhado',
                                        'a Liquidar': 'empenhos_R$_a_liquidar',
                                        'Liquidado': 'empenhos_R$_liquidado',
                                        'Pago': 'empenhos_R$_pago',
                                        'RP Inscrito': 'empenhos_RP_inscrito',
                                        'RP a Liquidar': 'empenhos_RP_a_liquidar',
                                        'RP Liquidado': 'empenhos_RP_liquidado',
                                        'RP Pago': 'empenhos_RP_pago'})

# retirar os pontos e virgulas da coluna cnpj e cpf deixar apenas os números
base_empenhos['empenhos_doc_fornecedor'] = base_empenhos['empenhos_doc_fornecedor'].str.replace('.', '', regex=False).str.replace('-', '', regex=False).str.replace('/', '', regex=False)

#base_empenhos = base_empenhos[base_empenhos['empenhos_num_empenho'].astype(str).str.startswith(('2024', '2025', '2026'), na=False)]

display(base_empenhos.head(2))
base_empenhos.info()
```
