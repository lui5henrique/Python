```python
import pandas as pd
import datetime
```

```python
ec = pd.read_excel('contratos.gov.xlsx', header=1, decimal=',', thousands='.')
display(ec.head())
ec.info()
```

```python
# droop colunas que não são necessárias
ec.drop(columns=['Fornecedor', 'Atualizado em', 'Data Encerramento'], inplace=True)

display(ec.head())
ec.info()
```

```python
# mudar nome das colunas
ec.rename(columns={'Número do instrumento': 'numero_instrumento',
                   'Vig. Início': 'vigencia_inicio',
                   'Vig. Fim': 'vigencia_fim',
                   'Valor Global': 'valor_global',
                   'Núm. Parcelas': 'numero_parcelas',
                   'Valor Parcela': 'valor_parcela'}, inplace=True)

display(ec.head())
ec.info()
```

```python
# trocar "Indeterminado" por data na coluna "vigencia_fim"
ec['vigencia_fim'] = ec['vigencia_fim'].replace('Indeterminado', '31/12/2026')

#trocar tipo de coluna "vigencia_inicio" e "vigencia_fim" para datetime
ec['vigencia_inicio'] = pd.to_datetime(ec['vigencia_inicio'], format='%d/%m/%Y')
ec['vigencia_fim'] = pd.to_datetime(ec['vigencia_fim'], format='%d/%m/%Y')

# trocar "Indetermindado" por valor 0 na coluna "numero_parcelas"
ec['numero_parcelas'] = ec['numero_parcelas'].replace('Indeterminado', 0)

display(ec.head())
ec.info()
```

```python
# mudar colunas R$_Valor_Global e R$_Valor_Parcela para float
ec['valor_global'] = (
    ec['valor_global']
    .astype(str)
    .str.replace(r'[R$\s.]', '', regex=True)  # remove R$, espaços e pontos
    .str.replace(',', '.', regex=False)       # troca vírgula por ponto
)
ec['valor_global'] = pd.to_numeric(ec['valor_global'], errors='coerce')
ec['valor_global'] = ec['valor_global'].fillna(0)

ec['valor_parcela'] = (
    ec['valor_parcela']
    .astype(str)
    .str.replace(r'[R$\s.]', '', regex=True)  # remove R$, espaços e pontos
    .str.replace(',', '.', regex=False)       # troca vírgula por ponto
)
ec['valor_parcela'] = pd.to_numeric(ec['valor_parcela'], errors='coerce')
ec['valor_parcela'] = ec['valor_parcela'].fillna(0)

display(ec.head())
ec.info()
```

```python
# calcular a diferença entre as datas "vigencia_fim" e "vigencia_inicio" e criar uma nova coluna "duracao_contrato" tipo Int representando a duração do contrato em meses
ec['duracao_contrato'] = (ec['vigencia_fim'] - ec['vigencia_inicio']).dt.days / 30
ec['duracao_contrato'] = ec['duracao_contrato'].astype(int)
display(ec.head())
```

```python
# fazer comparação entre a coluna "duracao_contrato" e "numero_parcelas" e criar uma lista mnostrando todos os registros que são diferentes
diferentes_parcelas = ec[ec['duracao_contrato'] != ec['numero_parcelas']]
display(diferentes_parcelas)
ec.info()
```

```python
# coluna valor_global dividir pela coluna duração do contrato e criar uma nova coluna "valor_parcela_calculado"
ec['valor_parcela_calculado'] = ec['valor_global'] / ec['duracao_contrato']

# fazer tratamento na coluna "valor_parcela_calculado" para arredondar para 2 casas decimais
ec['valor_parcela_calculado'] = ec['valor_parcela_calculado'].round(2)

# mostrar todos os registros da planilha
display(ec.head())
ec.info()
```

```python
# listar os registros onde a coluna "valor_parcela_calculado" é diferente da coluna "valor_parcela"
diferentes_valor_parcela = ec[ec['valor_parcela_calculado'] != ec['valor_parcela']]
display(diferentes_valor_parcela)

#listar os registros onde a coluna "valor_parcela_calculado" é igual a coluna "valor_parcela"
iguais_valor_parcela = ec[ec['valor_parcela_calculado'] == ec['valor_parcela']]
display(iguais_valor_parcela)
```

```python
# exportar em abas diferente mas na mesma planilha
with pd.ExcelWriter('erros_contratos_comparacao.xlsx') as writer:
    ec.to_excel(writer, sheet_name='Contratos_Reprogramado', index=False)
    diferentes_valor_parcela.to_excel(writer, sheet_name='Diferentes_Valor_Parcela', index=False)
    iguais_valor_parcela.to_excel(writer, sheet_name='Iguais_Valor_Parcela', index=False)
```
